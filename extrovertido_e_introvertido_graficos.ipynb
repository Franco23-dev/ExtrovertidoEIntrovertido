#Importaciones de librerias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Abrimos el dataset

```python
df = pd.read_csv("personality_datasert-copia.csv")
```

#Revisamos las columnas

```python
df.head()
```

#Cambiamos los nombres de las columnas en ingles a español

```python
print(df.columns)
español = dict(zip(list(df.columns), ['tiempo_que_pasa_solo','miedo_escenico', 'asistencia_a_eventos_sociales', 'salir_a_la_calle', 'agotado_despues_de_socializar', 'tamanio_de_circulo_de_amigos', 'frecuencia_de_publicaciones', 'personalidad']))
df = df.rename(español, axis=1)
print(df.columns)
```

#Histograma 1

```python
bins_publicaciones = np.arange(0, df['frecuencia_de_publicaciones'].max() + 2, 1)
sns.histplot(data=df, x='frecuencia_de_publicaciones', bins=bins_publicaciones)
```

#Histograma 2

```python
bins_tamanio_circulo_amigos = np.arange(0, df['tamanio_de_circulo_de_amigos'].max()+2 ,2)
sns.histplot(data=df, x='tamanio_de_circulo_de_amigos', bins=bins_tamanio_circulo_amigos, kde=True)
```

#Barplot

```python
sns.barplot(data=df,x='personalidad', y='frecuencia_de_publicaciones', color= sns.color_palette()[1])
```

#Scaterplot 1

```python
sns.scatterplot(data=df, x='salir_a_la_calle', y='tiempo_que_pasa_solo')
```

#Scaterplot 2

```python
sns.scatterplot(data=df, x='tamanio_de_circulo_de_amigos', y='tiempo_que_pasa_solo')
```

#Boxplot

```python
sns.boxplot(data=df,x='miedo_escenico', y='frecuencia_de_publicaciones', legend=False)
plt.title("Frecuencia en publicaciones segun el miedo escenico")
plt.xlabel("Miedo escenico")
plt.ylabel("Frecuencia en publicaciones")
plt.show()
```
